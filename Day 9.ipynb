## Chapter 8 Grouping vs Pivot Table vs Cross Tab

```python
insurance_data= pd.read_csv('insurance.csv')
```

```python
insurance_data
```

## Two types of data
-  discrete - countable,not measurable, can't be seggregated
-  continous - measurable,seggregated 

## 8.1 What is the average insurance charges based on different regions?

```python
insurance_data.groupby(by="region") ["charges"].mean().sort_values(ascending=False).round()
```

## 8.2 What is the average insurance charges based on gender?

```python
insurance_data.groupby(by="sex") ["charges"].mean().sort_values(ascending=False).round()
```

## 8.3 What is the average insurance charges based on gender and region?

```python
insurance_data.groupby(by=["region","sex"]) ["charges"].mean().sort_values(ascending=False).round()
```

## Pivot Table

```python
pd.pivot_table(data= insurance_data, values="charges", index=["sex","region"],aggfunc='mean').round()
```

```python
pd.pivot_table(data= insurance_data, values="charges", index=["region"], columns=["sex","smoker"],aggfunc='mean').round()
```

```python
insurance_data.aggregate()
```

## Cross Tab

```python
pd.crosstab(index=insurance_data.sex,columns=[insurance_data.region,insurance_data.smoker])
```
